```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
file1="E:\\Naresh IT\\Sample\\Visadataset.csv"
visa=pd.read_csv(file1)
visa
```

```python
data=visa['prevailing_wage']
round(np.percentile(data,25),2)
```

```python
plt.boxplot(data,vert=False)
plt.show()
```

```python
Q1=round(np.quantile(data,0.25),2)
Q2=round(np.quantile(data,0.50),2)
Q3=round(np.quantile(data,0.75),2)
IQR=(Q3-Q1)
LB=Q1-1.5*IQR
UB=Q3+1.5*IQR
print(Q1)
print(Q2)
print(Q3)
print(IQR)
print(LB)
print(UB)
```

```python
c1=data<LB
c2=data>UB
outlier=visa[c1|c2]
```

```python
(len(outlier)/len(visa))*100
```

```python
med_veg=visa['prevailing_wage'].median()
med_veg=round(med_veg,2)
med_veg
```

```python
og_wage=visa['prevailing_wage'].to_list()
out_wage=outlier['prevailing_wage'].to_list()
visa['prevailing_wage'].head()
```

```python
for i in out_wage:
    if i in og_wage:
        ind=og_wage.index(i)
        og_wage[ind]=med_veg
```

```python
visa['prevailing_wage']=np.where(visa['prevailing_wage']>UB,med_veg,visa['prevailing_wage'])
```
