```python
for i in continent:
    a1=visa[(visa['continent']==i)&(visa['case_status']=='Certified')]
    print(i,':',len(a1))
```

```python
for i in continent:
    con1=visa['continent']==i
    con2=visa['case_status']=='Certified'
    count1=len(visa[con1&con2])
    a1=visa[con1&con2]
    print('The number of Certified apilicatent in {} is:{}'.format(i,count1))
```

```python
for i in continent:
    con1=visa['continent']==i
    con2=visa['case_status']=='Denied'
    count1=len(visa[con1&con2])
    a1=visa[con1&con2]
    print('The number of Certified apilicatent in {} is:{}'.format(i,count1))
```

```python
certified,denied=[],[]
for i in continent:
    con1=visa['continent']==i
    con2=visa['case_status']=='Certified'
    con3=visa['case_status']=='Denied'
    count1=len(visa[con1&con2])
    count2=len(visa[con1&con3])
    certified.append(count1)
    denied.append(count2)
    
```

```python
pd.DataFrame(zip(certified,denied),columns=['Certified','Denied'],index=visa['continent'].unique())
```
