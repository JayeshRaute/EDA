```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.fillna(40,inplace=True)
df
```

```python
data={"Names":['Jay','rahul',np.nan,'raju'],
      'Age':[np.nan,23,22,21],
      'City':['Hyd','Blr','Pune',np.nan]}
df=pd.DataFrame(data)
df
```

```python
df['Age'].fillna(mean_age)
```
